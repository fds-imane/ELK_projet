### Conversion des bon types de données

```python
#Transformer les dates

# Pour les clients
customers['first_purchase'] = pd.to_datetime(customers['first_purchase'])
customers['last_purchase'] = pd.to_datetime(customers['last_purchase'])

# Pour les transactions
transactions['invoice_date'] = pd.to_datetime(transactions['invoice_date'])

print("Datetime now exploitables !")
```

### Analyse & doublon du dataset

```python
# Analyse rapide du dataset
datasets = {"Clients": customers, "Transactions": transactions, "Campagnes": campaigns}

for nom, df in datasets.items():
    print(f"\n--- Analyse : {nom} ---")
    print(f"Lignes : {df.shape[0]} | Colonnes : {df.shape[1]}")

    # Vérification des doublons pour verifier lunicité
    nb_doublons = df.duplicated().sum()
    if nb_doublons > 0:
        print(f"{nb_doublons} lignes en double détectées !")
    else:
        print("propre propre propre !!!!")
```

Hypothèse : il se pourrait que le système ai pu valider 2 fois la même ligne si l'utilisateur rafraîchit sa page ou s'il y a un bug réseau ce qui expliquerait les doublons. Pour ne pas prendre de risques, nous allons les supprimer afin d'éviter de surestimer le CA de l'entreprise.

```python
# Nettoyer les doublons et corriger les ids

# On supprime les 34 522 doublons des transactions
transactions = transactions.drop_duplicates()

# On convertit l id client en entier
transactions['customer_id'] = transactions['customer_id'].astype('Int64')

print(f" Doublons supprimés. Nouveau volume : {len(transactions)} lignes.")
```

### Valeurs manquantes

```python
# Checkpoint des valeurs manquantes en %
def check_missing(df):
    missing_pct = (df.isnull().sum() / len(df)) * 100
    # On ne garde que les colonnes qui ont des manques pour y voir clair
    return missing_pct[missing_pct > 0].round(2)

print("Pourcentage de valeurs manquantes :")
print(f"Clients :\n{check_missing(customers)}\n")
print(f"Transactions :\n{check_missing(transactions)}\n")
print(f"Campagnes :\n{check_missing(campaigns)}")
```

Hypothèse : Nous remarquons que plus de 22% de customer_id n'est pas reliée à un client identifié, ceci est énorme, il se peut qu'on relève des bug dans le trackind de l'id client au moment de la vente.

*   Nous remarquons aussi que product_name & quantity sont  faible en proportions mais c'est étrange à creuser !



Nous avons observés une quantité de 50 000 clients enregistrés dans les données clients. Le volume de transactions s'élève à 1 837 137 transactions sur une période allant de 2007 à 2011. C'est un dataset qui couvre 5ans d'activités.

```python
# Le contrôle final

print("Rapport de qualité après nettoyage des données :\n")

print("Vérification des types:")
check_types = {
    "Transactions (Date)": transactions['invoice_date'].dtype,
    "Transactions (ID)": transactions['customer_id'].dtype,
    "Clients (Date)": customers['first_purchase'].dtype
}
for info, dtype in check_types.items():
    print(f" - {info}: {dtype} ")

# Vérification des doublons
print(f"\n Doublons restants dans Transactions : {transactions.duplicated().sum()}")

# État des Valeurs Manquantes (NaN)
print("\n Valeurs manquantes :")
def missing_report(df, name):
    pct = (df.isnull().sum() / len(df)) * 100
    print(f" --- {name} ---")
    print(pct[pct > 0] if pct.sum() > 0 else "Aucun vide !")

missing_report(transactions, "Transactions")
missing_report(customers, "Clients")

# Aperçu final
print("\n Aperçu des 3 premières lignes de Transactions :")
display(transactions.head(3))
```

**1 Analyse et nettoyage de transactions.csv**

Les customer_id manquants: que représentent ces lignes ? Faut-il les supprimer ou les conserver ?

**Analyse** : Environ 22,8% des transactions n'ont pas d'identifiant client. Cela correspond généralement à des ventes "Invités" (Guest Checkout) ou des erreurs techniques lors de l'export du CRM.

**Décision** : Nous conservons ces lignes pour ne pas sous-estimer le chiffre d'affaires global et la saisonnalité. Cependant, elles seront exclues lors de la segmentation comportementale (RFM) car l'identité client est manquante.



```python
# Identification des manquants
missing_cust = df_trans['customer_id'].isnull().sum()
print(f"Transactions sans customer_id : {missing_cust} ({missing_cust/len(df_trans):.1%})")
```

Les quantity négatives: d'où viennent-elles ? Quel impact sur les agrégations futures ?

**Analyse** : Les quantités négatives proviennent des retours clients, des annulations de commandes ou des ajustements d'inventaire.

**Décision** : Elles doivent être conservées pour le calcul du CA net (vision comptable), mais elles doivent être isolées pour ne pas biaiser le calcul du panier moyen marketing.

```python
# Identification des retours
print(f"Nombre de lignes avec quantité < 0 : {(df_trans['quantity'] < 0).sum()}")
```

Les unit_price à zéro: anomalie ou cas métier légitime ?


**Analyse** : Un prix nul peut être légitime (cadeaux promotionnels, échantillons, garanties) ou une anomalie (données manquantes, erreurs d'interface).

**Décision** : Nous les conservons s'ils sont associés à des noms de produits réels, mais nous les documentons comme étant des opérations marketing sans valeur monétaire directe.

```python
# Aperçu des produits à 0€
print("Exemples de produits à prix nul :")
print(df_trans[df_trans['unit_price'] == 0]['product_name'].unique()[:10])
```

Les product_code atypiques: identifiez les codes non-produits (frais de port, ajustements comptables, etc.)

Calculez line_total = quantity × unit_price. Vérifiez la cohérence.


**Analyse** : Certains codes comme POST (frais de port) ou D (Discount) ne sont pas des produits physiques mais des lignes de service ou d'ajustement.

**Décision** : Nous créons la colonne line_total pour vérifier la cohérence. Les lignes avec des quantités NaN sont supprimées car elles empêchent tout calcul fiable.

```python
# Suppression des lignes sans quantité pour la cohérence
df_trans = df_trans.dropna(subset=['quantity'])

# Calcul de la ligne totale
df_trans['line_total'] = df_trans['quantity'] * df_trans['unit_price']

# Identification des codes de service (contiennent des lettres et ne sont pas des variantes)
atypical_codes = df_trans[df_trans['product_code'].str.contains('^[a-zA-Z]+$', na=False)]['product_code'].unique()
print(f"Codes atypiques identifiés : {atypical_codes}")
```
